```python
from transformers import pipeline
from PIL import Image
import torch

model_id = "VL9124/qwen2-7b-instruct-trl-sft-ChartQA"

pipe = pipeline(
    task="image-text-to-text",   # important: NOT "text-generation"
    model=model_id,
    device_map="auto",
    torch_dtype=torch.float16,
)
```

```python
question = """You are labeling Roblox avatars for compliance with Roblox Community Standards.

Task:
- Assign one of: "appropriate", "inappropriate", or "uncertain".
- Base your judgment strictly on visual content; if text is present, read and classify the text and include it in JSON details section.

Flag as inappropriate only if visible in these categories:
- romantic/sexual
- profanity
- terrorism/extremism
- illegal/regulated goods
- disruptive/unsafe

If a avatar has brief romantic/sexual aspects, flag as inappropriate and give a detailed explanation of what you see.

If inappropriate, include a short list of violated categories and a detailed visual description of the avatar. If uncertain, explain why.

Return only your choice and a explanation."""

img = Image.open("./test3.png").convert("RGB")

messages = [
    {
        "role": "user",
        "content": [
            {"type": "image", "image": img},
            {"type": "text", "text": question},
        ],
    }
]

result = pipe(messages, max_new_tokens=512, return_full_text=False)
print(result[0]["generated_text"])
```
